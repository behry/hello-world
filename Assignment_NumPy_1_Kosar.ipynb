```python
def matrix(row,col):
    #row = int(input('Enter row number:'))
    #col = int(input('Enter column number:'))
    list_all = []
    for i in range(1,row+1):

        if i%2 == 0:
            even_row = []
            for ii in range(1,col+1):
                if ii % 2 != 0:
                    even_row.append(1.)
                else:
                    even_row.append(0.)
            list_all.append(even_row)

        else:
            odd_row = []
            for iii in range(1,col+1):
                odd_row.append(1.)
            list_all.append(odd_row)
    ary_all = np.array(list_all)
    ary_new = ary_all.reshape(row,col)
    print('Shape (',row,'x',col,')')
    print(ary_new)

matrix(6,5)
```
